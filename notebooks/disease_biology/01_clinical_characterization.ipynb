```python
# Import necessary libraries
from pathlib import Path
import pandas as pd
from tabulate import tabulate
```

```python
# Root directory of the project
root_dir = Path.cwd().parent.parent

# Load TCGA KIRC clinical data
clinical_data = pd.read_csv(
    root_dir / "data" / "clinical" / "tcga_kirc_clinical.tsv",
    sep="\t"
)

# Display information about the clinical data
clinical_data.info()
```

```python
# Display the first few rows of the clinical data
print(tabulate(clinical_data.head(), headers='keys', tablefmt='grid'))
```

```python
# Create a summary table containing all string/categorical columns
summary = pd.DataFrame({
    "variable": clinical_data.select_dtypes(
        include=["object", "string"]
    ).columns
})

summary["categories"] = summary["variable"].apply(
    lambda c: clinical_data[c].dropna().unique().tolist()
)

summary
```
