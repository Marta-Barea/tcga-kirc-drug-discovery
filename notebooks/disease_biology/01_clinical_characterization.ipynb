## 1. Introduction

<div align="justify">

This Jupyter Notebook is part of the **Disease Biology** phase of the project *"Integrative Multi-Omics Analysis for Drug Discovery in Clear Cell Renal Cell Carcinoma (ccRCC)"*. Specifically, it addresses the **Clinical Data Characterization** stage, focusing on the exploration and assessment of the clinical data available in the **[TCGA-KIRC (Kidney Renal Clear Cell Carcinoma)](https://portal.gdc.cancer.gov/projects/TCGA-KIRC)** cohort. 

TCGA-KIRC is part of **The Cancer Genome Atlas (TCGA)** program, a landmark cancer genomics initiative jointly supported by the **[National Cancer Institute (NCI)](https://www.cancer.gov/)** and the **[National Human Genome Research Institute (NHGRI)](https://www.genome.gov/)**.

</div>

## 2. Objetives

<div align="justify">

The objectives of this notebook are to review and prepare the downloaded clinical datasets, assess data completeness and quality, and characterize the clinical features of the patient cohort. 

To achieve this, a comprehensive statistical analysis will be conducted, including descriptive summaries of demographic and pathological variables, identification of missing data patterns, and evaluation of key clinical attributes. In addition, survival analyses will be performed to investigate the prognostic relevance of selected clinical variables and provide insights into patient outcomes.
