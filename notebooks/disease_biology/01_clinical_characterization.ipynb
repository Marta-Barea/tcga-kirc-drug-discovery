<div align="justify">

As shown in the dataset summary, the clinical dataset comprises **537 observations**, each corresponding to an individual patient, and **27 variables** containing demographic, clinical, and patient identification information.

Regarding data types, the dataset consists of **22 categorical variables** stored as `string` and **5 numerical variables** stored as `float64`. The numerical variables primarily represent time-related measurements, such as days to birth, days to death, and days to last follow-up.

The dataset also contains missing values. While some variables, such as `demographic.days_to_death`, are expected to have missing entries because the information is only available for deceased patients, a substantial proportion of missing values is observed in the variables prefixed with `diagnoses.1.*` and `diagnoses.2.*`. This is a consequence of the data structure provided by the **GDC**, which stores all available diagnosis records for each patient. The `diagnoses.0.*` variables correspond to the primary diagnosis and are available for nearly all patients, whereas the `diagnoses.1.*` and `diagnoses.2.*` variables represent additional diagnosis records. Since only a small subset of patients has multiple diagnosis entries, these variables are mostly empty and therefore exhibit a high percentage of missing values.


</div>

### 5.2. Data preparation

<div align="justify">

Based on the previous inspection, the clinical dataset was preprocessed to obtain a clean and consistent set of variables suitable for downstream analyses. The complete preprocessing pipeline is implemented in [`src/preprocessing/prepare_clinical_data.py`](../../src/preprocessing/prepare_clinical_data.py).

The main preprocessing steps are summarized below:

- **Reorganizing the dataset.** The patient identifiers (`id` and `submitter_id`) were moved to the first columns of the dataset to facilitate subsequent integration with other data modalities.

- **Retaining only the primary diagnosis.** Since the **[GDC](https://portal.gdc.cancer.gov)** stores multiple diagnosis records using the prefixes `diagnoses.0.*`, `diagnoses.1.*`, and `diagnoses.2.*`, only the primary diagnosis (`diagnoses.0.*`) was retained. The remaining diagnosis records correspond to a small subset of patients and contain a high proportion of missing values. After filtering, the common `diagnoses.0.` prefix was removed from the column names to improve readability.

- **Transforming clinical variables.** Patient age at diagnosis was computed from `days_to_birth`, which is stored by TCGA as the number of days before birth and therefore requires negation. In addition, the variables `days_to_last_follow_up` and `days_to_death` were converted from days to years.

- **Creating survival variables.** Two new variables were generated to support survival analysis:
  - `event`, a binary variable indicating whether the patient was deceased (`1`) or censored (`0`).
  - `survival_time`, defined as the time to death for deceased patients and the time to the last follow-up for censored patients.

- **Handling missing values.** Different strategies were applied depending on the variable:
  - Patients with missing values in `survival_time` or `ajcc_pathologic_stage` were removed, as these variables are essential for the subsequent analyses.
  - Missing values in the TNM staging variables (`ajcc_pathologic_t`, `ajcc_pathologic_n`, and `ajcc_pathologic_m`) were replaced with the corresponding unknown categories (`TX`, `NX`, and `MX`).
  - Missing values in `tumor_grade` were replaced with the unknown category `GX`.
  - Missing values in `laterality` were imputed using the mode.
  - Missing values in `age_at_diagnosis` were imputed using the cohort median.

As a result of these preprocessing steps, the clinical dataset contains standardized variables, harmonized time measurements, and complete information for all variables required in the downstream survival analyses.

</div>

## 6. Statistical analysis

### 6.1. Descriptive statistics

#### 6.1.1. Measure of central tendency and dispersion

```python
# Measure of central tendency and dispersion
descriptive_stats = (
    prepared_clinical_data
    .drop(columns=['id', 'submitter_id'])
    .describe(include='all')
)

display(descriptive_stats)
```

<div align="justify"> 

From the descriptive analysis, the following observations were made:

- The cohort was predominantly **male**, representing **63.9%** of all patients.
- Most patients (**66.8%**) were **alive at the last follow-up**, while **33.2%** experienced the event of interest (death).
- The majority of tumors were classified as **M0 (79.1%)** and **Stage I (48.8%)**, indicating that most patients were diagnosed **without distant metastasis and at an early pathological stage**.
- The most frequent primary tumor category was **T1a (25.2%)**, corresponding to **small tumors (≤4 cm) confined to the kidney**, suggesting that localized tumors were the most common presentation.
- **Grade 2 (G2)** was the predominant histological grade (**42.4%**), indicating an **intermediate degree of tumor differentiation**.
- Tumors were slightly more common in the **right kidney (52.7%)** than in the left (47.3%), although the distribution was relatively balanced.
- The mean age at diagnosis was **61.1 ± 12.0 years**, which is consistent with the typical age of onset for renal cell carcinoma.
- The median observed survival time was **3.22 years**, with a mean of **3.64 ± 2.64 years**.
